```python
import pandas as pd
from pathlib import Path
```

```python
df1 = pd.read_csv('Resources/Monopoly_Alcohol_Imports.csv')
df2 = pd.read_csv('Resources/Merged Suicide Rates.csv')
# Rename 'Entity' to 'Country' in df2 for consistency
df2 = df2.rename(columns={'Entity': 'Country'})

# Merge the datasets on 'Country' (Location in df1 and Entity in df2) and 'Year'
merged_df = pd.merge(df1, df2, left_on=['Location', 'Period'], right_on=['Country', 'Year'], how='inner')

# Select only the required columns
final_df = merged_df[['Country', 'Year', 'Alcohol use disorders (%)', 'Dim1', 'Dim1ValueCode']]

# Display the final DataFrame
final_df
```

```python
final_df.to_csv('AlcoholImports.csv', index=False)
```
