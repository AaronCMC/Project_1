```python
import pandas as pd
from pathlib import Path
```

```python
df1 = pd.read_csv('Resources/Monopoly_Alcohol_Distribution.csv')
df2 = pd.read_csv('Resources/Merged Suicide Rates.csv')

# Convert 'Period' and 'Year' columns to numeric
df1['Period'] = pd.to_numeric(df1['Period'], errors='coerce')
df2['Year'] = pd.to_numeric(df2['Year'], errors='coerce')

# Filter both DataFrames for the years 2010-2017
df1_filtered = df1[(df1['Period'] >= 2010) & (df1['Period'] <= 2017)]
df2_filtered = df2[(df2['Year'] >= 2010) & (df2['Year'] <= 2017)]

# Merge the filtered DataFrames on 'Location' & 'Entity' and 'Period' & 'Year'
merged_df = pd.merge(df1_filtered, df2_filtered, left_on=['Location', 'Period'], right_on=['Entity', 'Year'], how='inner')

# Remove duplicates from the merged DataFrame
merged_df_cleaned = merged_df.drop_duplicates()
merged_df_cleaned = merged_df_cleaned.dropna(axis=1, how='all')
# Display the cleaned merged DataFrame
merged_df_cleaned
```

```python
merged_df_cleaned.to_csv('Acoholdistribution.csv', index=False)
```
