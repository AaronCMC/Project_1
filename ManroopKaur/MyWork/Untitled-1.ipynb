```python
import pandas as pd
```

```python
df1 = pd.read_csv('Resources/Mental health Depression disorder Data.csv')
```

```python
df1.head()
```

```python
df2 = pd.read_csv('Resources/Suicide Rates from 2011-2017.csv', encoding='ISO-8859-1',header=2)
```

```python
pd.set_option('display.max_columns', None)
df2.head()
```

```python
# Clean df1 if necessary
if df1.columns.nlevels > 1:
    df1 = df1.iloc[4:]  # Skip metadata rows if applicable
    df1.columns = df1.columns.droplevel(0)  # Drop first level of multi-index if needed
    df1.columns = df1.columns.droplevel(0)  # Drop second level of multi-index if needed

# Rename columns if they are not as expected
df1 = df1.rename(columns={'Location': 'Country', 'Period': 'Year'})
```

```python
# Inspect df2 column names
print(df2.columns)
```

```python
# Rename columns based on actual names
df2 = df2.rename(columns={
    'Location': 'Country',
    'Period': 'Year',
    'Age-standardized suicide rates (per 100 000 population)': 'Both sexes',
    'Age-standardized suicide rates (per 100 000 population).1': 'Male',
    'Age-standardized suicide rates (per 100 000 population).2': 'Female'
})

# Check the columns again to confirm renaming
print(df2.columns)


```

```python
# Print the columns to verify
print(df2.columns)

# Proceed only if the columns are present
if all(col in df2.columns for col in ['Country', 'Year', 'Both sexes', 'Male', 'Female']):
    df2 = df2[['Country', 'Year', 'Both sexes', 'Male', 'Female']]
else:
    print("Some columns are missing or have different names.")
```

```python
# Check the data types and unique values for the 'Year' column
print(df1['Year'].dtype)
print(df1['Year'].unique())

# Inspect the first few rows of df1 and df2
df1.head()
df2.head()
```

```python
df2_cleaned = df2.rename(columns={
    'Country': 'Country',
    'Both sexes': 'Both sexes',
    'Male': 'Male',
    'Female': 'Female'
})

# Remove any non-numeric entries in 'Year'
df2_cleaned['Year'] = pd.to_numeric(df2_cleaned['Year'], errors='coerce')

# Drop rows where 'Year' could not be converted
df2_cleaned = df2_cleaned.dropna(subset=['Year'])

# Convert 'Year' to integers
df2_cleaned['Year'] = df2_cleaned['Year'].astype(int)

# Filter rows for the years 2012-2017
df2_cleaned = df2_cleaned[df2_cleaned['Year'].between(2010, 2017)]

# Clean df1
# Ensure that 'Year' is cleaned and correctly formatted
df1_cleaned = df1.rename(columns={
    'Entity': 'Country',
    'Year': 'Year'
})

# Remove any non-numeric entries in 'Year'
df1_cleaned['Year'] = pd.to_numeric(df1_cleaned['Year'], errors='coerce')

# Drop rows where 'Year' could not be converted
df1_cleaned = df1_cleaned.dropna(subset=['Year'])

# Convert 'Year' to integers
df1_cleaned['Year'] = df1_cleaned['Year'].astype(int)

# Filter rows for the years 2012-2017
df1_cleaned = df1_cleaned[df1_cleaned['Year'].between(2010, 2017)]

# Merge the datasets on 'Country' and 'Year'
merged_df = pd.merge(df1_cleaned, df2_cleaned, on=['Country', 'Year'], how='inner')

merged_df
```

```python
merged_df.to_csv('output.csv', index=False)
```
